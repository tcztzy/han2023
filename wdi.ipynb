```python
import wbgapi as wb
import pandas as pd
import requests_cache

requests_cache.install_cache()
wdi = pd.read_parquet('wdi.parquet')
ssf = wb.region.members('SSF')
wdi = wdi[wdi['Country Code'].isin(ssf)]
wdi.drop(columns=wdi.drop(columns=[str(year) for year in range(1960, 2005)]), inplace=True)
```

```python
items = []
for db in wb.source.info().items:
    wb.db = db["id"]
    count = len([item for item in wb.economy.info(db=db["id"]).items if item["region"] == "SSF"])
    #wb.data.DataFrame('all', wb.region.members('SSF'), range(2005, 2021))
    items.extend([
        db | item | {"db_id": db["id"]} for item in wb.series.info().items
    ])
indicators = pd.DataFrame(items)
```
